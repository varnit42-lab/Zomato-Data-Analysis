## DATA ANALYSIS

### 1. How many restaurants offer online ordering?

```python
df["online_order"].value_counts()
```

- The number of restaurants offering online orders are 58

### 2. What is the average cost for 2 people?

```python
df["approx_cost(for two people)"].mean()
```

- Average cost for 2 people is approximately 418 rupees

###  3. How many restaurants marked rating 4 and above 4?

```python
count = 0 
for values in df["rate"]:
    if values >= 4:
            count = count+1

print (count) 
```

- The no of restaurants marked rating 4 and above 4 are 34

### 4 . What is the highest no of votes received by any restaurants?

```python
highest  = 0
for values in df["votes"]:
    if values > highest:
         highest = values
print(highest)
    
    
```

### 5. Find the average rating of restaurants in each listed type.

```python
df.groupby("listed_in(type)")["rate"].mean()
```

- The above are the average ratings for each listed type
